```python
# Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(0)

n_samples = 1000

x1 = np.random.normal(loc=0, scale=1, size=n_samples)
x2 = 0.5 * x1 + np.random.normal(loc=0, scale=1, size=n_samples)  
x3 = -0.2 * x1 + 0.8 * x2 + np.random.normal(loc=0, scale=1, size=n_samples)  

data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3
})
correlation_matrix = data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

p.line(x, y, legend_label="sin(x)", line_width=2)

output_notebook()
show(p)
```

```python
# Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

np.random.seed(0)
x = np.random.rand(100) * 10  
y = np.random.rand(100) * 10  

sizes = np.random.rand(100) * 20  
colors = ['#%02x%02x%02x' % (int(r), int(g), 150) for r, g in zip(50 + 2*x, 30 + 2*y)]  

df = pd.DataFrame({
    'x': x,
    'y': y,
    'sizes': sizes,
    'colors': colors
})

p = figure(title="Bokeh Scatter Plot with Customized Sizes and Colors", x_axis_label='X', y_axis_label='Y')

p.circle(x='x', y='y', size='sizes', color='colors', alpha=0.6, source=df)

output_notebook()
show(p)
```

```python
# Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Random Data", background_fill_color="#fafafa")

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.5)

p.y_range.start = 0
p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'
p.grid.grid_line_color="white"
output_notebook()

show(p)
```

```python
# Create a Bokeh heatmap using the provided dataset.
#data_heatmap = np.random.rand(10, 10)
#x = np.linspace(0, 1, 10)
#y = np.linspace(0, 1, 10)
#xx, yy = np.meshgrid(x, y)

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.layouts import column

# Generate random data for heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create Bokeh figure
p = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1),
           toolbar_location=None, tools="", plot_width=400, plot_height=400)

# Convert data to 1D arrays
xgrid, ygrid = np.meshgrid(x, y)
source = {'x': xgrid.ravel(), 'y': ygrid.ravel(), 'value': data_heatmap.ravel()}

# Define color mapper
colors = ['#d7191c', '#fdae61', '#ffffbf', '#a6d96a', '#1a9641']  # Define colors for heatmap
mapper = linear_cmap(field_name='value', palette=colors, low=data_heatmap.min(), high=data_heatmap.max())

# Plot heatmap using rect glyphs
p.rect('x', 'y', width=1/10, height=1/10, source=source,
       line_color=None, fill_color=mapper)

# Add color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Customize plot properties
p.title.align = 'center'
p.xaxis.axis_label = 'X'
p.yaxis.axis_label = 'Y'

# Display plot
output_notebook()
show(column(p))
```
